### 3.
(a) Squared bias decreases, variance increses, training error monotonically decreses, test error U-shape, irreducible error horizontal line.  

(b) As flexibility increases, the bias(made by abstracting real problem to relatively simple problem) decreases, and variance(subject to small changes) increases, irreducible error(literally irreducible error by random noise) remains constant, training error(decreased by perfectly fitting the training data), test error(irreducible error + squared bias + variance) initially decreases then increases.

### 5.
Flexible approach can deal with highly non-linear problems well.  
Large n and small p (When data density is high)  
Less flexible approach would be preferred when the problem is simply linear or the sample of data is very small.

### 6.
Parametric approach is a model-based approach, which assumes a function's form and only change their parameters.  
Non-parametric approach is a model-free approach, which doesn't assume any functional form. ***+ It does contain parameters, but the number of parameters depends on the amount of data, and it is for model complexity, not a form.***
Parametric approach has an advantage that we can make the complex problem simpler. It's disadvantage is that it is less flexible than non-parametric approach, and might have very different functional form from the real relationship.
